```python
import matplotlib.pyplot as plt
import numpy as np

def f(t):
    x = 3 * np.cos(2 * np.pi * t * t) + 10 * np.cos(2 * np.pi * 10 * t) + 30 * np.cos(4 * np.pi * 10 * t)
    return x

T = 1
n = 100
t = np.linspace(0,T,n)
t

```

```python
dt = t[1] - t[0]
dt
```

```python
freqs = np.linspace(0,50,400)
freqs
```

```python
f_res=[]

for freq in freqs:
    omega = 2 * np.pi * freq
    res = np.sum((f(t) * np.exp(-1j * omega * t)) * dt)
    f_res.append(res)

f_res = np.array(f_res)
f_mag = 2 * np.abs(f_res)
plt.plot(freqs, f_mag, color='black')
plt.show()
```
